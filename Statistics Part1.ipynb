```python
## Open iris DataSet:

df=pd.read_csv('iris.csv')
df.head()
```

```python
## Ploting pairplot using seaborn.
## We have seen that it will  Plot pairwise relationships in a dataset.

sns.pairplot(df)
```

```python
## set hue='species'

sns.pairplot(df,hue="species")
```

### Calculating mean,median,mode:

* for mode: *SciPy* is a scientific computation library that uses NumPy underneath.

* SciPy stands for Scientific Python. It provides more utility functions for optimization, stats and signal processing. Like NumPy, SciPy is open source so we can use it freely.

```python
## numpy has inbuilt function as mean:
## We are calculating mean of 'sepal_length'

np.mean(df['sepal_length'])
```

```python
print(np.mean(df['sepal_length']))
print(np.median(df['sepal_length']))


from scipy import stats
print(stats.mode(df['sepal_length']))
```

```python
## Since there is no much difference between mean(5.843333) and median(5.8), we can say there are no outliars.

## But
## Suppose I appened a large value explicitely then mean value changes and we can say yes there are outliars:


## For this, I have to first convert the 'sepal_length' into list and then apply append.
```

```python
ls=list(df['sepal_length'])
ls.append(100) ## We have added 100 (as outliar) explicitely.


print(np.mean(ls)) #Printing the mean of new appended list
                   ## WE can see now mean changed from previous mean
```

```python
# But what if we take median of the appended list:
# It will give same value as orignal one. We have seen this concept in discussion:


print(np.median(ls)) # We can see median is same as mean
```

## Que. is how do we plot the 'bell curve' :

* For ploting it we have seen in seaborn plot called 'distplot'

```python
sns.FacetGrid(df,hue='species').map(sns.distplot,'sepal_length')
```

```python
## We can increase the size of plot by giving size parameter as:

sns.FacetGrid(df,hue='species',size=6).map(sns.distplot,'sepal_length')
```

```python
## I'm not able to see what color belong to what thing: i will add---> add_legend()


sns.FacetGrid(df,hue='species',size=6).map(sns.distplot,'sepal_length').add_legend()
```

#### Now by adding the add_legend we can clearly see:

* the blue color is for setosa
* the orange color is for versicolor
* the green color is for verginica

### Now go to pairplot we can see for sepal_length:

* maximum time orange and green color are overlapping while blue color not overlapping at all.

* We can see that there are 3 bell curved and yes they are normally distributed ie. sepal_length is normally ditributed.

* Can I convert ND to SND (Standard normal distribution): YES , with *mean=0 and variance=1* using **z-score fomrula.

### Ques is why at y-axis values are coming as 0.0,0.2,0.4,.....??

* As we can see the bell curve is our entire pdf(probability density function) 
* y-axis taking frequency values

```python
## we can also draw histogram for 'sepal_length':
## Use distplot and bydefault hist=True, therefore no need to give hist=True. 

sns.distplot(df['sepal_length'])
```

### We can see the bell curve is also coming,How to remove it: Just add a parameter kde=False

* Because Kde means: Whether to plot a *gaussian kernel density estimate* (kde).


* Now we will get only histogram after setting kde=False;
* What is Histogram: Histogram basically finds out the frequency. (for eg. in below graph: how many values are between 4.3 to 4.6--> 10 values (ie frequency=10),  Similarly from 4.7 to 5.3 (i'm just taking assumption) --> 30 values.)

* these all are noting but beans(for eg. 4.3 to 4.7 is one bean on x-axis) in histogram


### What is pdf:
* pdf is nothing but smoothen graph on top of Histogram

```python
sns.distplot(df['sepal_length'],kde=False)
```

```python
## Again if you add kde=True or just remove kde=False from above code It is bydefault taking true and making bell curve.

sns.distplot(df['sepal_length'])
```
